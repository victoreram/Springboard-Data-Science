#### Extracting Relevant Columns

```python
### Text columns to be used for later features
text_columns = [('album', 'comments'), ('album', 'information'), ('album', 'tags'),
                ('artist', 'comments'), ('artist', 'associated_labels'), ('artist', 'related_projects'), ('artist', 'tags'),
               ('track', 'comments'), ('track', 'genres_all'), ('track', 'information'), ('track', 'language_code'), 
                ('track', 'tags'), ('track', 'title')]

### Basic columns for data analysis
basic_columns = [('album','date_released'), ('album', 'title'), ('album', 'id'),
                ('artist', 'id'), ('artist', 'name'), ('artist', 'active_year_begin'), 
                 ('artist', 'active_year_end'), 
                 ('track', 'genres'), ('track', 'genres_top'), ('track', 'genres_all'),
                 ('track', 'title'), ('track', 'date_recorded')]

### Columns that should be in datetime format
dt_COLUMNS = [('track', 'date_recorded'),
            ('album', 'date_released'),
              ('artist', 'active_year_begin'),('artist', 'active_year_end')]

### Set tracks df to just basic columns
tracks_basic = tracks[basic_columns]

### Remove rows with blank genres
tracks_basic = tracks_basic[~tracks_basic[('track','genres_all')].isnull()]

### Remove rows with genres that don't have genre codes, which are mislabeled. 
tracks_basic.drop(tracks_basic[tracks_basic[('track', 'genres_all')].str.contains('[a-z]')].index, inplace=True)

### Replace blank top genres with 'NONE'
tracks_basic[('track', 'genres_top')].fillna('NONE')

### Assign datetime columns to datetime type
for column in dt_COLUMNS:    
    times = pd.to_datetime(tracks_basic[column], errors='coerce')
    years = [time.year for time in times]
    tracks_basic.loc[:,column] = years
    
    # Assume that any artist that doesn't have an active year end is still active
    if (column == ('artist', 'active_year_end')) and :
        
        tracks_basic[column].fillna(pd.to_datetime('today').year, inplace=True)
tracks_basic.head(15)
```

#### Next Goal: Categorize years by decade
Goal:
1. Extract year from tracks_basic
    - **NOW:** Use ('album', 'date_released') column
    - **LATER:** Add column ('track', 'release_year'), a unified column with a track's release year, then 'guess' release_year if not present; Drop the other year bins
    
2. Place year into bins

Categories:
- 0: No data
- 1: 2010's
- 2: 2000's
- 3: 1990's


etc.

```python
def categorize_year(year):
    if 2010 <= year <= 2018:
        year = 1
    elif 2000 <= year <= 2009:
        year = 2
    elif 1990 <= year <= 1999:
        year = 3
    elif 1980 <= year <= 1989:
        year = 4
    elif 1970 <= year <= 1979:
        year = 5
    elif 1960 <= year <= 1969:
        year = 6
    else:
        year = 0
    return year

year_columns = [('album', 'date_released'), ('artist', 'active_year_begin'), ('artist', 'active_year_end'), 
               ('track', 'date_recorded')]
tracks_basic[('track', 'release_year')] = [categorize_year(y) for y in tracks_basic[('album', 'date_released')]]
tracks_basic[('track', 'release_year')].astype('category')
tracks_basic.drop(columns=year_columns, inplace=True)
tracks_basic.info()
```
